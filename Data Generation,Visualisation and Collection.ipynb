```python
mean=np.array([0.0,0.0])
cov=np.array([[5,0],[0,5]])

mean2=np.array([5.0,6.0])
cov2=np.array([[1.5,0.2],[0.2,1.2]])

dist=np.random.multivariate_normal(mean,cov,500)

dist2=np.random.multivariate_normal(mean2,cov2,500)

plt.scatter(dist[:,0],dist[:,1])

plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```

```python
plt.scatter(X1[:,0],X1[:,1],label="Guava",color='g')
plt.scatter(X2[:,0],X2[:,1],label="Orange",color="orange")
plt.xlabel("X1(size)")
plt.ylabel("X2(sweetness)")
plt.legend()
plt.show()

dist=np.vstack((X1,X2))
print(dist[2,0])
```
